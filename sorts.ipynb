```python
from typing import List


def sort_words(words: List[str]) -> List[str]:
    words.sort()
    return words[::-1]

def sort_numbers(numbers: List[int]) -> List[int]:
    numbers.sort()
    return numbers[::-1]

def sort_decimals(numbers: List[float]) -> List[float]:
    numbers.sort()
    return numbers[::-1]

# do not modify below this line
print(sort_words(["cherry", "apple", "blueberry", "banana", "watermelon", "zucchini", "kiwi", "pear"]))

print(sort_numbers([1, 5, 3, 2, 4, 11, 19, 9, 2, 5, 6, 7, 4, 2, 6]))

print(sort_decimals([3.14, 2.82, 6.433, 7.9, 21.555, 21.554]))
```

```python
from typing import List

def get_word_length(word: str) -> int:
    return len(word)


def sort_words(words: List[str]) -> List[str]:
    words.sort(reverse=True, key=get_word_length)
    return words


def sort_numbers(numbers: List[int]) -> List[int]:
    numbers.sort(key=abs, reverse=True)
    return numbers


# do not modify below this line
print(sort_words(["cherry", "apple", "blueberry", "banana", "watermelon", "zucchini", "kiwi", "pear"]))

print(sort_numbers([1, -5, -3, 2, 4, 11, -19, 9, -2, 5, -6, 7, -4, 2, 6]))
```

 This function accepts a list of numbers and returns a new list of numbers sorted based on their absolute value, in ascending order. Hint: You may use the abs() function to get the absolute value of a number.
