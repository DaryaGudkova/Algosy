## С двух сторон

- дан один отсортированный массив
- палиндром
- ответ формируется за счет сужения области с двух сторон


Найти позиции двух чисел в отсортированном массиве, которые дают в сумме таргет

```python
def for_sum(l: list[int], target: int) -> list[int]:
    q1 = 0
    q2 = len(l)-1
    while q1<q2:
        if l[q1] + l[q2] > target:
            q2 -= 1
        elif l[q1] + l[q2] < target:
            q1 += 1
        elif l[q1] + l[q2] == target:
            return [q1,q2]
    return [-1,-1]

for_sum([-2,1,6,9,12,21], 18)
# Время: О(n)
# Память: О(1)
```

## Каждому по указателю

- дано несколько массивов или строк
- нужно искать объединение/пересечение и тд у двух последовательностей

Даны два отсортированных массива чисел. Вернуть все их общие элементы.

```python
def common(l1: list[int], l2: list[int]) -> list[int]:
    q1, q2 = 0, 0
    res = []
    while q1<len(l1) and q2<len(l2):
        if l1[q1] == l2[q2]:
            res.append(l1[q1])
            q1 += 1
            q2 += 1
        elif l1[q1] < l2[q2]:
            q1 += 1
        else:
            q2 += 1
    return res

common([0,1,2,3], [1,3,4])
# Время: O(n+m)
# Память: О(min(n,m))
```

Даны два отсортированных массива. Нужно объединить их в один отсортированный массив

```python
def concat(l1: list[int], l2: list[int]) -> list[int]:
    q1, q2 = 0, 0
    res = []
    while q1<len(l1) and q2<len(l2):
        if l1[q1] == l2[q2]:
            res.append(l1[q1])
            res.append(l2[q2])
            q1 += 1
            q2 += 1
        elif l1[q1] < l2[q2]:
            res.append(l1[q1])
            q1 += 1
        else:
            res.append(l2[q2])
            q2 += 1
    
    while q1 < len(l1):
        res.append(l1[q1])
        q1 += 1

    while q2 < len(l2):
        res.append(l2[q2])
        q2 += 1

    return res

concat([0,1,2,3], [1,3,4])
```

## Окно фиксированной длины

- дана переменная, обозначающая размер окна
- требуется работать с подряд идущими элементами

<img src="fixlenwindow.png" width="400">

**Пример.** Дан массив и число k. Вернуть максимальную сумму k подряд идущих элементов.


[3,2,0,9,1,2,8,5,2], k=5 -> 25

```python
def maxsum_k(nums: list[int], k: int) -> int:
    window_sum = 0
    for i in range(k):
        window_sum += nums[i]

    max_sum = window_sum

    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i-k]
        if window_sum > max_sum:
            max_sum = window_sum

    return max_sum

maxsum_k([3,2,0,9,1,2,8,5,2], 5)

# Время O(n)
# Память O(1)
```

**Пример.** Дан массив чисел и число k. Найти максимальное произведение k подряд идущих элементов

[3,0,5,9,4,1] k=3 -> 180

```python
#мой неоптимальный вариант
def muxmult_k(nums: list[int], k: int) -> int:
    window_mult = 0
    j=0
    #поиск первого ненулевого произведения за O(nk)
    while window_mult==0 and j+k-1<len(nums):
        mult = 1
        for i in range(j,j+k):
            mult *= nums[i]
        j+=1
        window_mult = mult

    max_mult = window_mult

    for i in range(j+k-1, len(nums)):
        if nums[i] != 0 and nums[i-k] != 0:
            window_mult = window_mult // nums[i-k] * nums[i]
            max_mult = max(max_mult, window_mult)
    return max_mult

muxmult_k([3,0,5,9], 1)
# Время O(nk)
# Память O(1)
```

```python
def maxmult_k(nums: list[int], k: int) -> int:
    # Инициализация первого окна
    zero_count = 0
    window_mult = 1
    
    for i in range(k):
        if nums[i] == 0:
            zero_count += 1
        else:
            window_mult *= nums[i]
    
    max_mult = 0 if zero_count > 0 else window_mult

    # Сдвигаем окно
    for i in range(k, len(nums)):
        out_elem = nums[i - k]   # элемент, который выходит из окна
        in_elem = nums[i]        # элемент, который входит в окно

        # Удаляем влияние выходящего элемента
        if out_elem == 0:
            zero_count -= 1
        else:
            window_mult //= out_elem

        # Добавляем входящий элемент
        if in_elem == 0:
            zero_count += 1
        else:
            window_mult *= in_elem

        # Вычисляем произведение текущего окна
        current_mult = 0 if zero_count > 0 else window_mult
        max_mult = max(max_mult, current_mult)

    return max_mult

muxmult_k([3,0,5,1,9,5,0,2,0,0], 3)
# Время O(n)
# Память O(1)
```

## Непересекающиеся окна

- нужно работать с подряд идущими группами элементов
- один элемент принадлежит одной группе (группы не пересекаются)

**Пример.** Дан отсортированный по возрастанию массив уникальных целых чисел. Сжать его, объединяя подряд идущие числа:
если числа идут подряд (отличаются на 1), то записываем их через стрелку, остальные пишем отдельно

[1,2,3,5,8,9,14]

["1->3", "5", "8->9", "14"]

```python
def ziparr(nums: list[int]) -> list[str]:
    left = 0
    right = 0
    result = []

    while left < len(nums):
        #расширяем правую границу
        while right + 1 < len(nums) and nums[right+1] == nums[right] + 1:
            right += 1
        #обрабатываем окно 
        if left == right:
            result.append(f"{nums[right]}")
        else:
            result.append(f"{nums[left]}->{nums[right]}")
        #переходим к следующей группе
        left = right + 1
        right = right + 1
    
    return result

ziparr([1,2,3,5,8,9,14]) 

# Время О(n)
# Память О(n)
```

**Пример.** Дан массив символов. Нужно сжать его, заменяя группы подряд идущих одинаковых символов на символ и количество повторений. Если символ встречается один раз, количество не записывается.

chars = ["x", "x", "y", "z", "z", "z"]

["x", "2", "y", "z", "3"]

```python
def compress_arr(chars: list[str])->list[str]:
    left, right = 0, 0 
    result = []

    while left < len(chars):
        while right < len(chars) and chars[right]==chars[left]:
            right += 1
        result.append(chars[left])
        if right-left > 1:
            result.append(str(right-left))

        left = right

    return result
compress_arr(["x", "x", "y", "z", "z", "z"])
# Время O(n)
# Память O(n)
```

## Пересекающиеся окна

- нужно найти самую длинную последовательность подряд идущих элементов с определенными свойствами и окна пересекаются

**Пример.** Дан массив состоящий из нулей и единиц, а также число k. Разрешается заменить не более k нулей на единицы. Нужно найти максимальную длину подряд идущей последовательности из единиц с учетом таких замен

```python
def maxones(arr: list[int], k: int) -> int:
    l = 0
    r = -1
    res = 0
    zeros = 0
    while l < len(arr):
        while r + 1 < len(arr) and (zeros < k or arr[r+1]==1):
            r += 1
            if arr[r] == 0:
                zeros += 1
           
        res = max(res, r - l + 1)

        if arr[l] == 0:
            zeros -= 1

        l += 1
        
    return res

maxones([1,0,1,0,1,0,1,1],2) #res=6

# Время O(n)
# Память O(1)
```

**Пример.** Дана строка s и число k. Найдите самую длинную подстроку, содержащую не более k уникальных символов, выведите ее длину

s = 'aabacbebebe' k = 3

7

```python
from collections import defaultdict

def unic(s: str, k: int) -> int:
    l = 0
    r = 0
    dict_unic = defaultdict(int)
    res = 0

    while l < len(s):
        while r < len(s) and (len(dict_unic) < k or s[r] in dict_unic):
            dict_unic[s[r]] += 1
            r+=1
        res = max(r-l, res)

        dict_unic[s[l]] -= 1
        if dict_unic[s[l]] == 0:
            del dict_unic[s[l]]

        l += 1

    return res

unic('aabacbebebe', 3)

# Время O(n)
# Память O(1)
```

```python
def palindrom(s: str) -> bool:
    chars = {}
    for i in s:
        chars[i] = chars.get(i,0)
    odd=0
    for val in chars.values():
        if val % 2 != 0:
            odd += 1
            if odd > 1:
                return False
    return True
palindrom('aabbcc')
# Время O(n)
# Память O(k)
```

Дана строка. Нужно отсортировать ее символы по убыванию частот.

BABBCBC -> BBBBCCA

```python
def sortcharfreq(s: str) -> str:
    freq_dict = {}
    for c in s:
        freq_dict[c] = freq_dict.get(c,0) + 1
    
    freq_list = [[] for _ in range(len(s))]    
    for char, freq in freq_dict.items():
        freq_list[freq].append(char)

    res = ''
    for i in range(len(s)-1, 0, -1):
        for j in freq_list[i]:
            res += j*i
    return res
sortcharfreq('BABBCBC')

# Время O(n)
# Память O(n)
```

[[2,4],[1,3],[5,6],[6,7]] -> [[1,4],[5,7]]

```python
def merge_str(a:list[int], b: list[int]) -> list[int]:
    return [min(a[0],b[0]), max(a[1],b[1])]

def intersect_str(a:list[int], b: list[int]) -> bool:
    return max(a[0],b[0]) <= min(a[1],b[1])

def strange(segments: list[list[int]]) -> list[list[int]]:
    segments.sort()

    res = [segments[0]]

    for i in range(1, len(segments)):
        if intersect_str(segments[i], res[-1]):
            res[-1] = merge_str(segments[i], res[-1])
        else:
            res.append(segments[i])

    return res

strange([[2,4],[1,3],[5,6],[6,7]]) 

# Время O(nlog(n))
# Память O(n)
```

## Два указателя на отрезках

- встречаешь несколько списков интервалов

<img src="twopointers_segments.png" width="400">

Даны два отсортированных списка интервалов. Найти все пересечения интервалов из первого и второго списков. 

segments1 = [[1,2],[3,4],[6,7]]
segments2 = [[0,5],[7,8]]

res = [[1,2],[3,4],[7,7]]

```python
def is_overlapping(a: list[int], b: list[int]) -> bool:
    return max(a[0],b[0])<=min(a[1],b[1])

def overlap_two(a: list[int], b: list[int]) -> list[int]:
    return [max(a[0],b[0]), min(a[1],b[1])]

def intersect(s1: list[list[int]], s2: list[list[int]]) -> list[list[int]]:
    p1, p2 = 0, 0
    res = []

    while p1<len(s1) and p2<len(s2):
        if is_overlapping(s1[p1], s2[p2]):
            res.append(overlap_two(s1[p1], s2[p2]))

        if s1[p1][1] < s2[p2][1]:
            p1 += 1
        else:
            p2 += 1
    
    return res

segments1 = [[1,2],[3,4],[6,7]]
segments2 = [[0,5],[7,8]]
intersect(segments1, segments2)

# Время O(n+m)
# Память O(max(n,m))
```

Даны расписания двух людей и минимальная длительность встречи k. Найти первый свободный слот размером k, когда оба могут встретиться

busy1 = [[1,2],[3,4],[7,9]]

busy2 = [[0,3],[7,8]]

```python
def is_overlapping_k(a: list[int], b: list[int], k) -> bool:
    return min(a[1],b[1]) - max(a[0],b[0]) >= k

def overlap_two_k(a: list[int], b: list[int], k) -> list[int]:
    return [max(a[0],b[0]), max(a[0],b[0])+k]

def findfree(s1: list[list[int]], s2: list[list[int]], k) -> list[int]:
    p1 = 0
    p2 = 0
    res = []

    while p1<len(s1)-1 and p2<len(s2)-1:
        f1 = [s1[p1][1], s1[p1+1][1]]
        f2 = [s2[p2][1], s2[p2+1][1]]
        if is_overlapping_k(f1,f2,k):
            return overlap_two_k(f1,f2,k)
        if f1[1] <= f2[1]:
            p1+=1
        else:
            p2+=1

    return res

busy1 = [[1,2],[3,4],[7,9]]
busy2 = [[0,3],[7,8]]
k = 2
res = [4,6]

findfree(busy1, busy2, k)

# Время O(n+m)
# Память O(1)
```

```python
def count_meeting_rooms(s: list[list[int]]) -> int:
    points = []
    for i in s:
        points.append([i[0],1])
        points.append([i[1],-1])

    points.sort()

    max_rooms = 0
    curr_rooms = 0
    for p in points:
        curr_rooms += p[1]
        max_rooms = max(max_rooms, curr_rooms)

    return max_rooms

count_meeting_rooms([[2,5],[1,3],[2,4]])

# Память O(n)
# Время O(n*log(n))
```

## Базовый бинарный поиск

Дан массив чисел отсортированный по возрастанию. Найти последнее вхождение элемента target и вернуть позицию или -1, если элемента нет.

[1,2,8,8,9,10,11,12,14]

target = 8

res = 3

```python
def search_last(arr: list[int], target: int) -> int:
    l, r = 0, len(arr)

    while r - l > 1:
        middle = l + (r-l)//2
        if arr[middle] > target:
            r = middle
        else:
            l = middle

    return l if arr[l] == target else -1

search_last([1,2,8,8,9,10,11,12,14], 8)

# Память O(1)
# Время O(log(n))
```

```python
def search_first(nums: list[int], target: int) -> int:
    l, r = -1, len(nums)-1

    while r-l>1:
        m = l + (r-l)//2
        if nums[m] >= target:
            r = m
        else:
            l = m
    return r if nums[r] == target else -1

def search_second(nums: list[int], target: int) -> int:
    l, r = 0, len(nums)

    while r-l>1:
        m = l + (r-l)//2
        if nums[m] > target:
            r = m
        else:
            l = m
    return l if nums[l] == target else -1

def search_range(nums: list[int], target: int) -> int:
    return [search_first(nums,target), search_second(nums,target)] 

search_range([2,7,7,7,7,8,9],7)

# Память O(1)
# Время O(log(n))
```

Дана последовательность из скобок 4 видов. Проверить что каждая открывающая скобка имеет соответствующую закрывающую скобку того же вида. в правильном порядке

<img src="stek.png" width="300">

```python
def skob(seq: list[chr]) -> bool:
    dict_sk = {
        "(" : ")",
        "{" : "}",
        "<" : ">",
        "[" : "]"
    }

    stack = []

    for i in seq:
        if i in dict_sk.keys():
            stack.append(i)
        else:
            if len(stack) == 0 or dict_sk[stack.pop()] != i:
                return False
    
    return len(stack) == 0

skob(['(', '('])

# Время O(n)
# Память O(n)
```

РЕШЕНИЕ: 
Читаем выражение слева направо по токенам (числам и операторам).

Если токен — число → кладём его в стек.

Если токен — оператор (+, -, *, /) →

- Достаём из стека два верхних числа (сначала правое, потом левое).

- Применяем оператор.

- Результат кладём обратно в стек.

В конце — на стеке остаётся одно число, это ответ.

```python
def pols_cow(tokens: str) -> float:
    stack = []

    operators = {
        '+': lambda a, b: a + b,
        '-': lambda a, b: a - b,
        '*': lambda a, b: a * b,
        '/': lambda a, b: a / b,
    }

    for token in tokens:
        if token not in operators.keys():
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()
            res = operators[token](a, b)
            stack.append(res)

    return stack[0]

pols_cow(["2", "1", "+", "3", "*"])
```

Проходим по массиву слева направо:

- Пока стек не пуст И текущий элемент > элемента на вершине стека:

    - Достаём индекс из стека
    - Записываем для этого индекса результат = текущий элемент

- Добавляем текущий индекс в стек

В конце те, кто остались в стеке, не нашли большего → у них остаётся -1

```python
def next_greater(arr: list[int]) -> list[int]:
    stack = [arr[-1]]
    res = [-1] * len(arr)
    for i in range(len(arr)-2, -1, -1):
        while stack and stack[-1] <= arr[i]: 
            stack.pop()
        if stack:
            res[i] = stack[-1]
        stack.append(arr[i])
    return res

next_greater([2,5,1,3,0,4])
# Время O(n)
# Память O(n)
```

## Псевдо стек

- стек можно заменить на один или несколько счетчиков
- в стек складываются только одинаковые элементы
- важен лишь размер стека, а не его содержимое

```python
def psevdo_stack(tokens: str) -> bool:
    balance = 0
    for token in tokens:
        if token == '(':
            balance += 1
        else:
            balance -= 1
        if balance < 0:
            return False
    return balance == 0
psevdo_stack(')()()')
# Время O(n)
# Память O(1)
```

ПРИМЕР:
Даны строки s и t. Нужно вернуть true если строки будут одинаковыми после ввода в текстовый редактор. Символ # означает, что предыдущий символ нужно стереть, а если перед # отсутствует символ, то стирать ничего не нужно.

s = "ab#c"

t = "ad#c"

res = true

РЕШЕНИЕ
Для каждой строки держим указатель с конца:

если видим обычный символ — это кандидат на сравнение
если видим # — увеличиваем счетчик удалений
пока счетчик удалений больше 0, пропускаем символы и уменьшаем счетчик

Пусть:

i — указатель в s
j — указатель в t

Для s:

идем влево
если s[i] == '#', увеличиваем skip
если это буква, но skip > 0, значит она должна быть удалена → пропускаем ее
если это буква и skip == 0, значит это настоящий символ для сравнения

То же самое делаем для t.

После этого:

если в обеих строках нашли символы, сравниваем их
если символы разные → false
если в одной строке символ есть, а в другой уже нет → false
если дошли до конца обеих строк одновременно → true

```python
def psevdo_stack(s: str, t: str) -> bool:
    s_pointer=len(s)-1
    t_pointer=len(t)-1
    while s_pointer >= 0 or t_pointer >= 0:
        s_skip = 0
        while s_pointer >= 0:
            if s[s_pointer] == '#':
                s_skip += 1
                s_pointer -= 1
            elif s_skip > 0:
                s_pointer -= 1
                s_skip -= 1
            else:
                break

        t_skip = 0
        while t_pointer >= 0:
            if t[t_pointer] == '#':
                t_skip += 1
                t_pointer -= 1
            elif t_skip > 0:
                t_pointer -= 1
                t_skip -= 1
            else:
                break
        
        if t_pointer >= 0 and s_pointer >= 0:
            if s[s_pointer] != t[t_pointer]:
                return False
        elif t_pointer >= 0 or s_pointer >= 0:
            return False
        s_pointer -= 1
        t_pointer -= 1
    return True
psevdo_stack("cde#ab###", "cd#")
# Время O(n)
# Память O(1)
```

```python
# Реализация префиксного массива
class PrefixArray:
    #подсчет префиксных сумм
    def __init__(self, nums: list[int]):
        self.prefix = [0]
        for num in nums:
            self.prefix.append(self.prefix[-1] + num)
    # сумма на отрезке
    def sum(self, l: int, r: int) -> int:
        return self.prefix[r + 1] - self.prefix[l]
```

<img src="lad.png" width="300">

```python
def max_sum(board: list[list[int]]) -> int:
    n = len(board)
    m = len(board[0])

    rows_sum = [0]*n
    cols_sum = [0]*m
    for i in range(n):
        for j in range(m):
            val = board[i][j]
            rows_sum[i] += val
            cols_sum[j] += val

    max_sum = 0
    for i in range(n):
        for j in range(m):
            curr_sum = rows_sum[i] + cols_sum[j] - board[i][j]
            if i == 0 and j == 0:
                max_sum = curr_sum
            max_sum = max(max_sum, curr_sum)
    return max_sum

# Время: O(n+m)
# Память: O(n+m)
```

```python
def pivot_index(nums: list[int]) -> int:
    total_sum = sum(nums)
    prefix_sum = 0

    for i, num in enumerate(nums):
        suffix_sum = total_sum - prefix_sum - num
        if prefix_sum == suffix_sum:
            return i
        prefix_sum += num
    return -1

# Время: O(n)
# Память: O(1)
```

## Базовые операции с односвязными списками

<img src="linked_list.png" width="250">

```python
# Создание односвязного списка 1-2-3
class Node:
    def __init__(self, data, next):
        self.data = data
        self.next = next
head = Node(1, Node(2, Node(3, None)))
```

```python
# Подсчет размера списка

def size(head: Node) -> int:
    count = 0
    curr = head
    while curr:
        count += 1
        curr = curr.next
    return count

# Время: O(n)
# Память: O(1)
```

```python
# Поиск середины списка

# медленный указатель двигается на один
# быстрый - на два элемента
# продолжаем пока быстрый указатель и следующий 
# за ним элемент не равны пустому значению

def middle(head: Node) -> Node:
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow
# Время: O(n)
# Память: O(1)
```

```python
# Разворот (без создания нового списка)

def reverse(head: Node) -> Node:
    prev = None
    curr = head
    while curr:
        tmp = curr
        curr = curr.next
        tmp.next = prev
        prev = tmp
    return prev
# Время: O(n)
# Память: O(1)
```

```python
def merge(head1: Node, head2: Node) -> Node:
    dummy = Node(0, None)
    curr = dummy

    while head1 or head2:
        if head2 is None or (head1 is not None and head1.data <= head2.data):
            curr.next = head1
            head1 = head1.next
        else:
            curr.next = head2
            head2 = head2.next
        curr = curr.next
    return dummy.next

def print_list(head: Node):
    current = head
    while current is not None:
        print(current.data, end = ' -> ')
        current = current.next
    print('None')

l1 = Node(1,Node(2,Node(3,None)))
l2 = Node(2,Node(2,Node(7,Node(8,None))))
print_list(l1)
print_list(l2)
print_list(merge(l1,l2))

# Время: O(n+m)
# Память: O(1)
```

Дана голова односвязного списка. Нужно проверить является ли он палиндромом.

Находим середину списка и разворачиваем правую часть списка, чтобы можно было идти с конца списка к середине и сравнивать соответствующие симметричные элементы паттерном 'с двух сторон'

Серединный узел после разворота будет указывать на пустой узел.

<img src="rev.png" width="250">

```python
def middle_node(head: Node) -> Node:
    fast, slow = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow

def reverse_list(head: Node) -> Node:
    curr = head
    prev = None
    while curr:
        tmp = curr
        curr = curr.next
        tmp.next = prev
        prev = tmp
    return prev

def is_palindrom(head: Node) -> bool:
    first_half_end = middle_node(head)
    second_half_begin = reverse_list(first_half_end)
    p1, p2 = head, second_half_begin
    while p2:
        if p1.data != p2.data:
            return False
        p1 = p1.next
        p2 = p2.next
    return True

is_palindrom(Node(1,Node(2,Node(2,Node(1,None)))))
# Время: O(n)
# Память: O(1)
```

## Брутфорс

- нужно сгенерировать все возможные комбинации, перестановки или подмножества без ограничений

<img src="bruteforce.png" width="350">

Дана непустая строка, содержащая цифры от 1 до 9. Каждая цифра соответствует буквам на телефоне. Нужно вернуть все возможные комбинации букв

<img src="tel.png" width="350">

```python
from collections import deque

def generate_combinations(s: str) -> list[str]:
    phone_map = {
        "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
        "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
    }

    result = deque([""])
    
    while result and len(result[0]) < len(s):
        prefix = result.popleft()
        digit = s[len(prefix)]

        for letter in phone_map[digit]:
            result.append(prefix + letter)

    return list(result)

# Время: O(n*4^n)
# Память: O(n*4^n)
```

примечание про deque:

deque — это двусторонняя очередь из модуля collections.

deque используют, когда нужно эффективно работать с очередью или стеком, особенно если часто требуется удалять элементы из начала. В обычном списке pop(0) работает медленно, потому что элементы сдвигаются, а в deque удаление слева выполняется быстро d.popleft() .

## Бэктрекинг

- нужно сгенерировать последовательности с определенными правилами

<img src="backtr.png" width="350">

Дано число n. Нужно сгенерировать все правильные скобочные последовательности из n скобок одного типа.
<img src="ck.png" width="350">

Идея генерировать сразу только правильные скобочные последовательности, а не все и потом отсеивать валидные.

```python
from collections import deque

def generate_brackets(n: int) -> list[str]:
    q = deque([("", 0, 0)])

    while q and len(q[0][0]) < 2 * n:
        prefix, open_used, close_used = q.popleft()

        if open_used < n:
            q.append((prefix + "(", open_used + 1, close_used))

        if close_used < open_used:
            q.append((prefix + ")", open_used, close_used + 1))
    
    return [prefix for prefix, _, _ in q]

generate_brackets(3)

# Для оценки сложности достаточно сказать количество всех 
# комбинаций и проговорить что на самом деле генерировать 
# будем не все, а отсечем их по некоторому правилу 

# Время приблизительно:  O(n*2^(2*n))
# Память приблизительно: O(n*2^(2*n))
```

```python
def count_nodes(node: TreeNode) -> int:
    if node is None:
        return 0
    left = count_nodes(node.left)
    right = count_nodes(node.right)

    return left + right + 1
```

## Снизу вверх

- нужно собирать ответ от листьев к корню
- ответ в каждом узле зависит от ответов детей

Дан корень бинарного дерева. Найти его диаметр (длину самого большого пути между двумя узлами)

<img src="tree6.png" width="200">

Считаем снизу вверх для каждого узла длину самой длинной ветки.
<img src="tr.png" width="250">

```python
def find_diameter(node: TreeNode) -> int:
    diameter = 0
    
    # возвращает высоту текущего поддерева: 
    # берём большую высоту из левого и правого поддерева 
    # и добавляем 1 за текущий узел
    def dfs(node: TreeNode) -> int:
        if node is None:
            return 0
        
        left = dfs(node.left)
        right = dfs(node.right)

        nonlocal diameter
        diameter = max(diameter, left + right)

        return max(left, right) + 1
    
    dfs(node)
    return diameter

# Время: O(n) n - число узлов в дереве
# Память: O(h) - высота дерева (память тратится при рекурсивных вызовах)
```

**Замыкание** — это механизм, при котором внутренняя функция сохраняет доступ к переменным внешней функции даже после своего создания. В данном коде dfs является замыканием, потому что использует и изменяет переменную diameter, объявленную в find_diameter.

## Сверху вниз

- нужно передавать дополнительную информацию (сумму, максимум, границу, номер уровня и тп) от корня к листьям для формирования ответа 

<img src="upd.png" width="400">

Дан корень бинарного дерева и число targetSum. Определить существует ли путь от корня до листа сумма значений которого равна targetSum.

<img src="ex3.png" width="200">
<img src="ex4.png" width="200">

```python
def has_path_sum(self, root: TreeNode, target_sum: int) -> bool:
    def dfs(node: TreeNode, curr_sum: int) -> bool:
        if node is None:
            return False
        
        curr_sum += node.val

        if node.left is None and node.right is None:
            return curr_sum == target_sum
        
        return dfs(node.left, curr_sum) or dfs(node.right, curr_sum)
    
    return dfs(root, 0)

# Время: O(n) n - число узлов в дереве
# Память: O(h) - высота дерева (память тратится на стек рекурсивных вызовов)
```

```python
# Построение графа их списка ребер 

# Важно: для неориентированного графа 
# каждое ребро добавляется дважды!

from collections import defaultdict

# неориентированный
def graph_from_edges(edges: list[list[int]]) -> dict[int, list[int]]:
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    return graph

# ориентированный
def graph_from_edges(edges: list[list[int]]) -> dict[int, list[int]]:
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
    return graph
```
