## С двух сторон

- дан один отсортированный массив
- палиндром
- ответ формируется за счет сужения области с двух сторон


Найти позиции двух чисел в отсортированном массиве, которые дают в сумме таргет

```python
def for_sum(l: list[int], target: int) -> list[int]:
    q1 = 0
    q2 = len(l)-1
    while q1<q2:
        if l[q1] + l[q2] > target:
            q2 -= 1
        elif l[q1] + l[q2] < target:
            q1 += 1
        elif l[q1] + l[q2] == target:
            return [q1,q2]
    return [-1,-1]

for_sum([-2,1,6,9,12,21], 18)
# Время: О(n)
# Память: О(1)
```

## Каждому по указателю

- дано несколько массивов или строк
- нужно искать объединение/пересечение и тд у двух последовательностей

Даны два отсортированных массива чисел. Вернуть все их общие элементы.

```python
def common(l1: list[int], l2: list[int]) -> list[int]:
    q1, q2 = 0, 0
    res = []
    while q1<len(l1) and q2<len(l2):
        if l1[q1] == l2[q2]:
            res.append(l1[q1])
            q1 += 1
            q2 += 1
        elif l1[q1] < l2[q2]:
            q1 += 1
        else:
            q2 += 1
    return res

common([0,1,2,3], [1,3,4])
# Время: O(n+m)
# Память: О(min(n,m))
```

Даны два отсортированных массива. Нужно объединить их в один отсортированный массив

```python
def concat(l1: list[int], l2: list[int]) -> list[int]:
    q1, q2 = 0, 0
    res = []
    while q1<len(l1) and q2<len(l2):
        if l1[q1] == l2[q2]:
            res.append(l1[q1])
            res.append(l2[q2])
            q1 += 1
            q2 += 1
        elif l1[q1] < l2[q2]:
            res.append(l1[q1])
            q1 += 1
        else:
            res.append(l2[q2])
            q2 += 1
    
    while q1 < len(l1):
        res.append(l1[q1])
        q1 += 1

    while q2 < len(l2):
        res.append(l2[q2])
        q2 += 1

    return res

concat([0,1,2,3], [1,3,4])
```

Дан массив чисел. Без создания нового массива переместить все нули в конец текущего массива.

```python
#мое решение(без паттерна "быстрый и медленный")
def inplace_zero(l: list[int]) -> list[int]:
    q0 = 0
    q1 = len(l)-1
    while l[q0] != 0:
        q0 += 1
    while l[q1] == 0:
        q1 -= 1

    while q0 < q1:
        #print(q0,q1)
        print(l)
        if l[q0] != 0:
            q0 += 1
        else:
            for i in range(q0, q1):
                l[i] = l[i+1]
            l[q1] = 0
            q1 -= 1
    return l

inplace_zero([0,1,0,0,3,4,2])
```

```python
# мое решение c паттерном "быстрый и медленный"

def inplace_zero(l: list[int]) -> list[int]:
    slow = 0 #указывает на позицию куда ставить то что нашел быстрый указатель
    fast = 0 #ищет следующий ненулевой элемент
    while fast < len(l):
        if l[fast] == 0:
            fast += 1
        else:
            temp1 = l[fast]
            temp2 = l[slow]
            l[slow] = temp1
            l[fast] = temp2
            slow += 1
            fast += 1
    return l

inplace_zero([0,1,0,0,3,4,2])
```

```python
# аналогичное решение из видео
# c паттерном "быстрый и медленный"

def inplace_zero(l: list[int]) -> list[int]:
    slow = 0 #двигается по условию
    fast = 0 #всегда двигается
    while fast < len(l):
        if l[fast] != 0:
            l[slow], l[fast] = l[fast], l[slow]
            slow += 1
        fast += 1
    return l

inplace_zero([0,1,0,0,3,4,2])
# Время О(n)
# Память O(1)
```

```python
# решение через 2 указателя
def del_blank(l:list[str]) ->list[str]:
    prev_space = False #флаг о том что предыдущий элемент был пробелом
    slow = 0
    for fast in range(len(l)):
        if l[fast] == ' ':
            if not prev_space:
                l[slow] = l[fast]
                slow += 1
            prev_space = True
        else:
            l[slow] = l[fast]
            slow += 1
            prev_space = False
    del l[slow:]
    return l

del_blank([' ', ' ', 'h', 'i', ' ', ' ', ' ', '!', ' '])
```
