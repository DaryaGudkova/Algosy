```python
# учусь писать класс односвязного списка

#TODO: написать везде типы 
class Node:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def push_front(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    def push_back(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

    def push_pos(self, value, position):
        new_node = Node(value)

        if position == 0:
            self.push_front(value)
        else:
            current = self.head
            k = 1
            while k < position:
                current = current.next
                k += 1
            new_node.next = current.next
            current.next = new_node
    
    def print_list(self):
        current = self.head
        while current is not None:
            print(current.value, end = ' -> ')
            current = current.next
        print('None')

    def find(self, value):
        current = self.head
        while current is not None:
            if current.value == value:
                return True
            current = current.next
        return False
    
    def find_by_pos(self, position):
        current = self.head
        k = 0
        while k < position-1:
            current = current.next
            k += 1
        return current.value

    
    def remove(self, value):
        if self.head is None:
            return
        
        if self.head.value == value:
            self.head = self.head.next
            return
        
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next

    def remove_by_pos(self, position):
        if position == 1:
            self.head = self.head.next
        else:
            current = self.head
            k = 0
            while k < position-2:
                current = current.next
                k += 1
            current.next = current.next.next


    def reverse_list(self):
        prev = None
        current = self.head

        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        self.head = prev         

lst = LinkedList()
lst.push_front(70)
lst.push_back(10)
lst.push_front(80)
lst.push_back(90)
lst.print_list()

lst.remove(10)
lst.print_list()

lst.push_back(40)
lst.push_back(50)

lst.print_list()
lst.reverse_list()
lst.print_list()

lst.push_pos(3,1)
lst.print_list()

print(lst.find_by_pos(1))
lst.remove_by_pos(1)
lst.print_list()
```

```python
l = LinkedList()

q = int(input())

for _ in range(q):
    parts = input().split()
    query_type = int(parts[0])

    if query_type == 1:
        x = int(parts[1])
        y = int(parts[2])
        l.push_pos(y,x)
    elif query_type == 2:
        x = int(parts[1])
        print(l.find_by_pos(x))
    else:
        x = int(parts[1])
        l.remove_by_pos(x)
```
